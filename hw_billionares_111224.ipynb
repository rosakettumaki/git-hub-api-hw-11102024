```python
## 4a) How many male billionaires are there compared to the number of female billionares?

#male_billionaires = df[df['gender'] == 'male']
#female_billionaires = df[df['gender'] == 'female']

male_billionaires = df[df['gender'] == 'male']['name'].count()

female_billionaires = df[df['gender'] == 'female']['name'].count()

print( male_billionaires)
print(female_billionaires)
```

```python
## 4b) What percent is that?
# https://www.geeksforgeeks.org/how-to-calculate-the-percentage-of-a-column-in-pandas/
#https://stackoverflow.com/questions/67650853/how-to-get-the-percentage-of-two-values-of-a-column-given-every-row-has-either


#df[percent] = (df['column_name'] / df['column_name'].sum()) * 100.round(2)

pcts = df['gender'].value_counts() / df['gender'].count() * 100

print('% male', pcts['male'])
print('% female', pcts['female'])
```

```python
## 4c) Is the average wealth different for a female billionaire as compared to a male billionaire?


#male_billionaires = df[df['gender'] == 'male']['name'].count()

male_billionaires_w = df[df['gender'] == 'male']['networthusbillion'].mean().round(2)

female_billionaires_w = df[df['gender'] == 'female']['networthusbillion'].mean().round(2)

print('Male billionaire average wealth', male_billionaires_w)
print('Female billionaire average wealth', female_billionaires_w)
```
